```python
import pandas as pd
import glob
import os

# Read all metrics CSV files in the current folder
csv_files = glob.glob(os.path.join(os.getcwd(), "*.csv"))
metrics_dfs = {os.path.basename(f): pd.read_csv(f) for f in csv_files}

# Statistical analysis for each file
analysis_results = {}
for fname, df in metrics_dfs.items():
    stats = df.describe().T  # basic statistics
    stats['std'] = df.std()
    stats['median'] = df.median()
    stats['min'] = df.min()
    stats['max'] = df.max()
    stats['mean'] = df.mean()
    stats['file'] = fname
    analysis_results[fname] = stats

# Display analysis for each file
for fname, stats in analysis_results.items():
    print(f"Analysis for {fname}:\n")
    display(stats)
    print("\n")

# Optionally, aggregate across all files for overall statistics
all_metrics = pd.concat(metrics_dfs.values(), ignore_index=True)
overall_stats = all_metrics.describe().T
overall_stats['std'] = all_metrics.std()
overall_stats['median'] = all_metrics.median()
overall_stats['min'] = all_metrics.min()
overall_stats['max'] = all_metrics.max()
overall_stats['mean'] = all_metrics.mean()
print("Overall statistics across all files:\n")
display(overall_stats)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style for seaborn
sns.set(style="whitegrid")

# 1. Boxplots for RMSE, MAE, R2, and infer_ms by model
fig, axes = plt.subplots(1, 4, figsize=(24, 5))
metric_names = ['rmse', 'mae', 'r2', 'infer_ms']
model_names = [os.path.splitext(os.path.basename(f))[0] for f in csv_files]

for i, metric in enumerate(metric_names):
    data = []
    for model, df in metrics_dfs.items():
        for value in df[metric]:
            data.append({'Model': os.path.splitext(model)[0], metric: value})
    plot_df = pd.DataFrame(data)
    sns.boxplot(x='Model', y=metric, data=plot_df, ax=axes[i])
    axes[i].set_title(f'{metric.upper()} by Model')
    axes[i].set_xlabel('Model')
    axes[i].set_ylabel(metric.upper())

plt.tight_layout()
plt.show()

# 2. Summary Table: Mean and Std for each metric by model
summary = []
for model, df in metrics_dfs.items():
    summary.append({
        'Model': os.path.splitext(model)[0],
        'RMSE Mean': df['rmse'].mean(),
        'RMSE Std': df['rmse'].std(),
        'MAE Mean': df['mae'].mean(),
        'MAE Std': df['mae'].std(),
        'R2 Mean': df['r2'].mean(),
        'R2 Std': df['r2'].std(),
        'Infer_ms Mean': df['infer_ms'].mean(),
        'Infer_ms Std': df['infer_ms'].std(),
    })
summary_df = pd.DataFrame(summary)
display(summary_df)

# 3. Highlight best model(s) for each metric
best_rmse = summary_df.loc[summary_df['RMSE Mean'].idxmin()]
best_mae = summary_df.loc[summary_df['MAE Mean'].idxmin()]
best_r2 = summary_df.loc[summary_df['R2 Mean'].idxmax()]
best_infer = summary_df.loc[summary_df['Infer_ms Mean'].idxmin()]

print("Best RMSE model:", best_rmse['Model'], "Mean RMSE:", best_rmse['RMSE Mean'])
print("Best MAE model:", best_mae['Model'], "Mean MAE:", best_mae['MAE Mean'])
print("Best R2 model:", best_r2['Model'], "Mean R2:", best_r2['R2 Mean'])
print("Best inference speed model:", best_infer['Model'], "Mean infer_ms:", best_infer['Infer_ms Mean'])
```

```python
import itertools
import numpy as np

# ---- 1. collect per-fold RMSE for each model -----------------------------
model_rmse = {m: df.sort_values('fold')['rmse'].values
              for m, df in metrics_dfs.items()}

model_names = list(model_rmse)
n_folds     = len(next(iter(model_rmse.values())))  # assumes same length

# ---- 2. exact paired-permutation test ------------------------------------
def exact_paired_p(rmse_a, rmse_b):
    """Two-sided exact permutation p-value for paired data."""
    diffs      = rmse_a - rmse_b                      # per-fold signed diff
    obs_mean   = diffs.mean()                         # observed statistic
    signs      = np.array(list(itertools.product([-1, 1], repeat=len(diffs))))
    null_means = (signs * diffs).mean(axis=1)         # every flipped mean
    p = (np.abs(null_means) >= abs(obs_mean)).mean()  # two-tailed
    return obs_mean, p

# ---- 3. run all pairwise comparisons -------------------------------------
results = {}
for i, m1 in enumerate(model_names):
    for m2 in model_names[i+1:]:
        obs, p = exact_paired_p(model_rmse[m1], model_rmse[m2])
        results[(m1, m2)] = (obs, p)

# ---- 4. print nicely ------------------------------------------------------
for (m1, m2), (obs, p) in results.items():
    note = "significant" if p < 0.05 else "not significant"
    print(f"{m1} vs {m2}: RMSE = {obs:+.4f}, p = {p:.4f} ({note})")
```
